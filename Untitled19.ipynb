```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Création d'un ensemble de dates
dates = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")

# Simulation des prix d'actions (quotidiens)
stock_data = pd.DataFrame({
    'Date': dates,
    'Stock_Price': np.random.normal(loc=150, scale=20, size=len(dates))  # Prix aléatoires pour les actions
})

# Simulation des rendements des obligations (mensuels)
bond_dates = pd.date_range(start="2023-01-01", end="2023-12-31", freq="M")
bond_data = pd.DataFrame({
    'Date': bond_dates,
    'Bond_Return': np.random.normal(loc=0.002, scale=0.01, size=len(bond_dates))  # Rendements aléatoires pour les obligations
})

# Fusion des données d'actions et d'obligations
portfolio_data = pd.merge(stock_data, bond_data, on='Date', how='outer').sort_values(by='Date')

# Ajout de valeurs manquantes pour simuler des évaluations irrégulières
portfolio_data.loc[portfolio_data['Stock_Price'].isnull(), 'Stock_Price'] = np.nan
portfolio_data.loc[portfolio_data['Bond_Return'].isnull(), 'Bond_Return'] = np.nan

# Visualisation des données simulées
plt.figure(figsize=(12, 6))
plt.plot(portfolio_data['Date'], portfolio_data['Stock_Price'], label='Stock Price', marker='o')
plt.plot(portfolio_data['Date'], portfolio_data['Bond_Return'] * 1000, label='Bond Return (scaled)', marker='o')
plt.title('Simulated Financial Portfolio Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt

# Remove rows with missing values for interpolation
interpolation_data = portfolio_data.dropna()

# Points for interpolation
interpolation_points = pd.to_datetime(['2023-01-05', '2023-01-12', '2023-01-19', '2023-01-26'])

# Perform Lagrange interpolation for stock prices and bond returns
lagrange_interpolated_stock_prices = lagrange(interpolation_data['Date'].values, interpolation_data['Stock_Price'].values)(interpolation_points)
lagrange_interpolated_bond_returns = lagrange(interpolation_data['Date'].values, interpolation_data['Bond_Return'].values)(interpolation_points)

# Visualize the results
plt.figure(figsize=(12, 6))
plt.plot(portfolio_data['Date'], portfolio_data['Stock_Price'], label='Stock Price', marker='o')
plt.plot(portfolio_data['Date'], portfolio_data['Bond_Return'] * 1000, label='Bond Return (scaled)', marker='o')

# Plot the Lagrange interpolation curve
plt.scatter(interpolation_points, lagrange_interpolated_stock_prices, color='red', label='Lagrange Interpolation of Stock Prices')
plt.scatter(interpolation_points, lagrange_interpolated_bond_returns * 1000, color='green', label='Lagrange Interpolation of Bond Returns (scaled)')

plt.title('Simulated Financial Portfolio Data (One Month) with Lagrange Interpolation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
